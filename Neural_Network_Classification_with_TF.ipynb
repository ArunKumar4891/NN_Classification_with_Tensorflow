```python
# import the lib

import tensorflow as tf
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# create data to view and fit

from sklearn.datasets import make_circles


# make 1000 examples

n_samples =  1000


# create circles

x, y = make_circles(n_samples,
                   noise=0.03,
                   random_state=42)
```

#### steps in modeling 


#### 1. create or import the model
#### 2. compile the model

#### 3. fit the model

#### 4. tweak and evaluate


```python
# set the random seed

tf.random.set_seed(42)


#1 .create the model


model_1= tf.keras.Sequential([
    
    tf.keras.layers.Dense(1)
])


# 2. compile the model

model_1.compile(loss = tf.keras.losses.BinaryCrossentropy(),
               optimizer = tf.keras.optimizers.SGD(),
               metrics = ["accuracy"])     
# 3. fit the model

model_1.fit(x,y, epochs=100)



```

```python
model_1.fit(x, y, epochs=200, verbose = 0)

```

```python
model_1.evaluate (x,y)
```

```python
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])


model_2.compile(loss = tf.keras.losses.BinaryCrossentropy(),
               optimizer = tf.keras.optimizers.SGD(),
               metrics = ["accuracy"])


model_2.fit(x,y, epochs=100)
```

```python
model_2.evaluate (x,y)
```

```python

# improving our model


# set the random seed

tf.random.set_seed(42)


#1 .create the model


model_3= tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    
    tf.keras.layers.Dense(1)
])


# 2. compile the model

model_3.compile(loss = tf.keras.losses.BinaryCrossentropy(),
               optimizer = tf.keras.optimizers.Adam(),
               metrics = ["mae"])     




# 3. fit the model

model_3.fit(x,y, epochs=100)
```

```python
model_3.evaluate (x,y)
```

### find the best learning rate




```python
# set the random seed


tf.random.set_seed(42)


#1. create a model

model_9 = tf.keras.Sequential([
    
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(1, activation = "sigmoid"),
    
])


#2. compile the model

model_9.compile(loss = "binary_crossentropy",
                optimizer= "Adam",
                metrics = ["accuracy"])


# create the learning rate callback

lr_schedular = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20)
                                                       
                                                       )

#4 . fit the model 


model_9 = model_9.fit(x, y, epochs = 100, callbacks=[lr_schedular])

```

```python
""" Building a multiclass classification model"""


# for our multiclass classification model, we can use a similar arch 

# input shape = 28*28 (shape of the image)
# output shpae = 10 ( one per class of clothing)
# loss fun  = tf.keras.lossess.CategoricalCrossentropy()
#output layer activation = softmax (not sigmoid)

```

```python

train_data[0].shape
```

```python
# our data needs to be flattend 


flatten_model = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28, 28))])
```

```python
flatten_model.output_shape
```

```python
# set random seed

tf.random.set_seed(42)


# create a model


model_11 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(4, activation= "relu"),
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(10, activation = tf.keras.activations.softmax)
    
    
    
])



# compile the model

model_11.compile (loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                 optimizer = tf.keras.optimizers.Adam(),
                 metrics= ["accuracy"])


# fit the model

model_11.fit(train_data, train_labels, epochs = 10, validation_data=(test_data, test_labels))
```
